```python
import numpy as np
import matplotlib.pylab as plt
```

Since python interprets floats by default as double, we plug in $\varepsilon = 1E-16$ 
In our order of magnitude estimate, the formula for $\delta$ turned out to be $(\frac{f\varepsilon}{f^{(5)}})^{\frac{1}{5}}$

***Here we calculate delta, 
then use our central difference to differentiate***

***for an exponential function, the n-th derivative is the same as function value itself. Therefore, $\delta = (\varepsilon)^{\frac{1}{5}}$



```python
eps = 1e-16

dx = eps**(1/5)

def four_pt_cdf(f, x, dx):
    return (f(x - 2 * dx) + 8 * f(x + dx) - 8 * f(x - dx) - f(x + 2 * dx))/(12 * dx)
```

First we compute the numerical error using our estimated dx. This turned out to be ~$10^{-12}$.

Compute, on a log-log scale, the error in comparison to 'analytical' value. Our estimated dx is ploted in a vertical line. We see that it is indeed on the same order of magnitude

# First we do the calculation for f=Exp[x]

```python
xx = np.logspace(0, 1.4,500)
ddx = np.logspace(-8,0, 1000)

plt.plot(ddx, np.abs(np.exp(10) - four_pt_cdf(np.exp, 10, ddx)), 'ko', ms = 1)
plt.plot()
plt.xscale('log')
plt.yscale('log')
plt.axvline( x = dx, c= 'r', label = "Estimated step size")
plt.title("Absolute Error in Differentiating Exp(x) \nwith Different dx for x = 10")
plt.xlabel('dx')
plt.ylabel('$\epsilon (f\')$')
plt.legend()
print(xx.max())
```

```python
err = np.abs(np.array([[np.exp(x) - four_pt_cdf(np.exp, x, delta) for delta in ddx] for x in xx]))
min_idx = np.argmin(err, axis = 1)
min_dx = np.take(ddx, min_idx)

plt.plot(xx, np.take(ddx, min_idx), 'ko', ms = 1)
plt.xlabel('x')
plt.ylabel('Optimal dx')
plt.yscale('log')
plt.xscale('log')
plt.axhline(y=dx, c= 'r', label = 'Estimated Optimal dx')

plt.axhline(y=np.mean(min_dx), c= 'b', label = 'Averaged Optimal dx', ls='--')
plt.legend()
```

```python
print('The average dx at which error is minimized is ', f'{np.mean(min_dx):.1E}')
print('Our estimated value of dx is ', f'{dx:.1E}')
```

Our estimated value is on the same order as computed optimal value.

# Now we proceed to do the calculation for $f=\text{Exp}[0.01x]$

```python
xx = np.logspace(0, 1.4,500)
ddx = np.logspace(-8,0, 1000)

plt.plot(ddx, np.abs(np.exp(.1) - four_pt_cdf(np.exp, .1, ddx)), 'ko', ms = 1)
plt.plot()
plt.xscale('log')
plt.yscale('log')
plt.axvline( x = dx, c= 'r', label = "Estimated step size")
plt.title("Absolute Error in Differentiating $Exp(0.01x)$ \nwith Different dx for x = 10")
plt.xlabel('dx')
plt.ylabel('$\epsilon (f\')$')
plt.legend()
print(xx.max())
```

```python
err = np.abs(np.array([[np.exp(.01* x) - four_pt_cdf(np.exp, .01*x, delta) for delta in ddx] for x in xx]))
min_idx = np.argmin(err, axis = 1)
min_dx = np.take(ddx, min_idx)

plt.plot(xx, np.take(ddx, min_idx), 'ko', ms = 1)
plt.xlabel('x')
plt.ylabel('Optimal dx')
plt.yscale('log')
plt.xscale('log')
plt.axhline(y=dx, c= 'r', label = 'Estimated Optimal dx')

plt.axhline(y=np.mean(min_dx), c= 'b', label = 'Averaged Optimal dx', ls='--')
plt.legend()
```

```python
print('The average dx at which error is minimized is ', f'{np.mean(min_dx):.1E}')
print('Our estimated value of dx is ', f'{dx:.1E}')
```

The estimate is slightly better for f=Exp(0.01x) compared to f=Exp(x). 

### The computed optimal dx seems to be x independent as well.

# In both cases, the estimated value is on the same order as the computed optimal.
