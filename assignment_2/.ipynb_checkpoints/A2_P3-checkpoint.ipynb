```python
import numpy as np
import matplotlib.pylab as plt
```

```python
chebfit = np.polynomial.chebyshev.chebfit
chebval = np.polynomial.chebyshev.chebval
```

$log_2(y(x))$ where $x\in (-1,1)$ requires that $y(x) = \frac{(x+3)}{4}$,

such that valuating $log_2(y(x))$ for $x\in(-1,1)$ is equivalent to evaluating $log_2(y)$ for $y\in(.5,1)$.

Below, we fit Chebyshev polynomials from 2 to 40 degrees at 100 points between -1 and 1. 
After that, we plot the maximum error within the 100 points agains the fitting order.

```python
trans = lambda x: (x+3)/4

## create points between -1 and 1
x = np.linspace(-1, 1, 1000)
## (linearly) map to (0.5, 1), take loog
y = np.log2(trans(x))

plt.plot(x,y)
```

### Plotting how error scales with Chebyshev polynomial degree

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.set_yscale('log')
ax.grid('minor')

d = []
err = []
est_err = []
for deg in np.arange(2,40):
    coef, diag = chebfit(x, y, deg = deg, full = True)
    red_res = np.sqrt(diag[0][0]/len(x))
    est_err.append(red_res)
    yy = chebval(x, coef)
    # xx
    d.append(deg)
    # computing the error
    max_err = np.max(np.abs(y-yy))
    err.append(max_err)

# Plotting error between chebfit and the true function.
plt.plot(d,err, 'ko', ms = 2)
plt.xlabel('Fitting Order')
plt.ylabel('Max error')
```

From the plot we see that the error terms drop below $10^{-6}$ beyond the $7$th order fit.

```python
xdeg = np.arange(0,41)
coef_40, diag = chebfit(x, y, deg = 40, full = True)
coef_40
```

We see that the coefficient drops to 1E-15 as we get to ~18-th order term when using Chebyshev polynomials up to fortieth degree. 

## Let's truncate everything less than $1E-8$ part of the largest coefficient. 

We multiply the array by boolean `cond` to truncate without changing size of the array.

```python
lim = 1e-8
cond = np.array([np.abs(c/coef_40.max()) >= lim  for c in coef_40])
coef = coef_40 * cond.astype(int)
coef
```

## NOTE that this is fitting $log(f(x))$ where f(x) is a transformation from (-1,1) to (0,5,1).

```python
plt.plot(trans(x), np.abs(chebval(x, coef) - np.log2(trans(x))), 'ko', ms = 1)
```

Let's figure out an algorithm for calculating $log_e(x)$. 

### This is equivalent to $\frac{log_2(x)}{log_2(e)}$. 

### Let's focus on the denominator. 

$e \approx 0.6795(mantissa)\cdot 2^{2(exponent)}$

Therefore, $log_2(e) = log_2(\text{mantissa})+log_2(2^2) = log_2(\text{mantissa})+\text{exponent}\times log_2(2) = log_2(\text{mantissa}) + \text{exponent}$. 

Similarly, we can decompose the numerator into log(mantissa) + exponent. The ratio would be equivalent to taking the natural log of (x).

*I did some searching and it seems that the mantissa has to be $\in (0.5,1)$ in python convention.*


```python
# creating positive flating numbers
x_test = np.logspace(1, 100,10000)

man_test, exp_test  = np.frexp(x_test)
# # log2(x) is the function we want to mimic using chebval
# plt.plot(x_test, np.log2(man_test) + exp_test, '--')
# ## chebval mimics log2(f(x)), if we use log2(f(f^-1(x))) then it should be the same
# plt.plot(x_test, chebval(4*man_test - 3, coef) + exp_test, '--')
# plt.plot(x_test, np.log2(x_test), 'r+', ms = .5)

plt.plot(x_test, man_test, 'ko', ms = .5)
plt.xscale('log')
plt.xlabel('x')
plt.ylabel('mantissa')
plt.title("Plotting Mantissa vs. $x$ Values")
```

### A caveat is that we need to first map the mantissa from $(0.5,1)$ to $(-1,1)$ since our Chebyshev polynomial was fitted in this range.
Previously we used $y = (x+3)/4$. The inverse (from (0.5, 1) to (-1,1)) is $x = 4*y-3$.

$log(f(x)) = log(f(f^{-1}(y))) = log(y)$

## QUESTION: 
the assignment asks for the natural log, but the function name was given as `mylog2`

```python
def mylog2(x, base = np.e):
    ## create a function that maps from (0.5, 1) to (-1, 1) for chebval
    inv = lambda y: y * 4 - 3
    
    ## break log_b(a) = log(a)/log(b) into mantissa + exponents
    man_1, exp_1 = np.frexp(x)
    man_base, exp_base = np.frexp(base)
    
    num = chebval(inv(man_1), coef) + exp_1

    den =  chebval(inv(man_base), coef) + exp_base
    
    return num/den
```

```python
man_base, exp_base = np.frexp(2)
man_base
np.log2(man_base)
```

```python
chebval(4*(man_base)-3, coef)
```

```python
plt.plot(x_test, mylog2(x_test, base = np.e), '--', dashes=(5, 5), label = 'Our Poly. fit')
plt.plot(x_test, np.log(x_test), zorder = -1, label = 'numpy\'s log')
plt.xscale('log')
plt.legend()
```

```python
plt.plot(x_test, np.abs(mylog2(x_test, base = np.e) - np.log(x_test))\
         , 'k.', ms = .1, label = 'Abs. Error')
plt.xscale('log')
plt.legend()
```

Seems like the error is pretty uniform and fluctuates with the mantissa (which is sent into our polynomial evaluation).
