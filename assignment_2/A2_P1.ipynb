### The problem can be setup in spherical coordinate with $(r,\theta,\phi)$. 
The shell integral method integrates $E(z,\theta,\phi)$ for the sphere of radius R at constant $\phi$ (on the x-y plane).
The full integral is $\iint_{\phi,\theta} E(z,\theta,\phi)\ d\theta\ d\phi$. 
First get rid of $\theta$ based on cylindrical symmetry. We pull the factors out:

$$\frac{1}{4\pi\varepsilon_{0}}(2\pi R^2\sigma) \int_{0}^{\pi} \frac{(z-Rcos\phi)sin\phi}{(R^2+z^2-2Rzcos\phi)^{3/2}} d\phi$$

We can make another coordinate transform (as shown on Griffith) into $u\in(-1,1)$. In this interval, orthorgonalities of Legendre polynomials is satisfied:

$$\frac{1}{4\pi\varepsilon_{0}}(2\pi R^2\sigma) \int_{-1}^{1} \frac{(z-Ru)}{(R^2+z^2-2Rzu)^{3/2}} d\phi$$


Let's focus on the integral part first and see what `scipy` does

```python
from scipy import integrate as itg
import numpy as np
import matplotlib.pylab as plt
```

```python
def elem(z, r, u):
    top = (z - r * u)
    bot = (r**2 + z**2 - 2*r*z*u)**(3/2)
    return top/bot
```

```python
R = 3
zz = np.linspace(-3 * R, 3*R, 100)
zz = np.sort(np.append(zz,[R,-R]))
```

```python
fig = plt.figure()
ax = fig.add_subplot()

integral = np.vectorize(lambda z: itg.quad(lambda u: elem(z,R,u), -1, 1))
val, err =  integral(zz)


ax.errorbar(zz, val, yerr = err, fmt = 'k|', ms = 6, label = 'scipy.integrate.quad', zorder = 5)
ax.axvline(R,c='r', ls = '--',lw =.5)
ax.axvline(-R,c='r', ls = '--', lw = .5)

ax.set_xlabel('$z$ Height')
ax.set_ylabel('$E_z$ strength')
```

For our integration routine, we will implement **Legendre polynomial fitting (reviewed Jon's code)**

NOTE: `legvander(x,order):`

    returns the pseudo-vandermonde matrix for legendre polynomials

, where row corresponds the $x$ and columns corresponds the polynomial $P_m$. $$M_{n,m}=P_{Legendre,m}(x_n)$$
Now we have the polynomial, we can solve the system 
$$y_n = M.c \\
        = \sum_{i=0}^{m}P_{Legendre,m}(x_n) \cdot c_m$$.

We only need $c_0$ due to orthogonality. Extracting this from the first row of inverse matrix:
$$c_0 = M^{-1}_{0,i} (y^T)^i $$



### In short: 

1. generate pseudo-vandermonde mat. by dividing (-1,1) into `order +1` intervals
    
**NOTE: the integral range doesn't need to be in (-1,1). We simply move every `order+1` points TO (-1,1), fit Legendre polynomials, and extract coefficient of $P_0$**
    
2. Use the first row of inverse matrix to fit the next `order+1` $y$ values in our set of $y$ and get $c_0$.
(fixing those y values in (-1,1))

3. $c_0 * \text{order} * dx$ is our estimate of the integral in that small region of f(x)

NOTE: we tailor the total number of points without compromising the `order` parameter.
    

```python
def legendre_int(func, xmin, xmax, order):
    ## generates our Legendre fitting interval
    x=np.linspace(-1,1,order+1)
    mat=np.polynomial.legendre.legvander(x,order)
    inv = np.linalg.inv(mat)
    ## Extract coefficient to get c0
    coef_0 = inv[0,:]

    ## target number of point (best effort)
    npt_tgt = 100
    ## intervals needed. 'order' number of points per interval
    n_int = npt_tgt//order
    ## how we actually divide our integration region
    npt_use =  order *n_int + 1
    
    xx = np.linspace(xmin, xmax, npt_use)
    dx = np.median(np.diff(xx))
    yy = func(xx)

    total = 0
    
    ## Integrating each order n interval by extrating Legendre order 0 coefficient
    for i in range(n_int):
        y = yy[i * order: (i + 1) * order+1]
        total += np.sum(coef_0 * y) * order * dx
    
    return total
```

```python
my_integral = np.vectorize(lambda z: legendre_int(lambda u: elem(z,R,u), -1, 1, 4))
my_val =  my_integral(zz)

ax.plot(zz, my_val, 'ro', ms = 2, label = 'Legendre Fitting', zorder = 10)
fig.legend()
fig
```

NOTE: our code does throw `Runtime Warning: dividing by zero` at $\pm R$. There is a visible singularity in the integral from the plot. The total electric field asymptotically blows up as $\lim z\rightarrow R$.

a) scipy.integrate.quad **did not** care, nor did it throw any error.

b) Our legendre fitting scheme warned us of dividing by zero. This is due `z` being on the surface and magnitude of the electric field $\sim 1/r^2$ and $r\rightarrow 0$.

c) The analytical solution is $\frac{2\pi R^2\sigma}{4\pi\varepsilon_0} \frac{2}{z^2}$. We only need to check against  the $\frac{2}{z^2}$ part.

## Checking the errors against true solution

```python
def field(z):
    # Notice this is R independent. Point charge approximation is valid.
    return 2/z**2


zz_err = np.linspace(-1.2*R, -.8*R, 200)
zz_err = np.append(zz_err, -R)

err = np.abs(integral(zz_err)[0] - field(zz_err))
my_err = np.abs(my_integral(zz_err) - field(zz_err))

plt.clf()
plt.yscale('log')
plt.plot(zz_err, err, 'k.',ms=1, label = 'scipy.integrate.quad',zorder = 10)
plt.plot(zz_err, my_err, 'ro', ms = 1, label = 'Legendre')
plt.legend()
```
