# Reparametrization
Expand the paraboloid equation:
$$ z-z_0=a x^2-2 a x_0 x+a x_0^2+a y^2+a y_0^2-2 a y y_0\\
z=a x^2+a y^2-2 a x x_0-2 a y y_0+\left(a x_0^2+a y_0^2+z_0\right)$$
Notice that if we group $z_0, a y_0^2, a x_0^2$ into a single constant term $D$, the problem becomes linear in $A,B,C = a, -2ax_0, -2a y_0$.

$$z=A x^2+A y^2+B x+C y+D$$
Now we have system with four degrees of freedom, with which we can perform linear fitting (**A is matrix not parameter here**):
$$(A^T N^{-1}A)^{-1}A^{T}N^{-1}y=m$$

```python
a = np.vstack([x**2 + y**2, x, y, np.ones(len(x))]).transpose()
print('Shape of A'.rjust(10), a.shape)
np.abs(a).max()/np.abs(a).min()
```

## Since our measurement of $z$ is given without uncertainty, we simply set $N=\mathbb{1}$.

```python
corr = np.linalg.inv(a.transpose()@a)
m = corr@a.transpose()@z
m
```

# Uncertainty of reparametrized fitted parameters ABCD

```python
plt.imshow(np.log(np.abs(corr)))
plt.colorbar()
print("Fitted parameters",m)
print("Uncertaintied", np.sqrt(np.diag(corr)))
print("Relative Error", np.sqrt(np.diag(corr))/m)
```

```python
zz = a@m
ax.plot_trisurf(x, y, zz, cmap='viridis', edgecolor='none', linewidths=0.2)
fig
```

# Convert back to original set of parameters
$$A=a\\
B=-2ax_0 \rightarrow x_0 = -\frac{B}{2A}\\
C=-2ay_0 \rightarrow y_0 = -\frac{C}{2A}\\
D=ax_0+ay_0+z_0 \rightarrow z_0 = D + A\frac{B}{2A}+A \frac{C}{2A} = D+\frac{B+C}{2}
$$

```python
m
A,B,C,D = m
aa = A
x0 = -B/2/A
y0 = -C/2/A
z0 = D + (B+C)/2
params = [aa, x0,y0,z0]
print("Best fit parameters are a, x0, y0, z0 = \n", params)
```

# Error on data

Error in the data can be computed from $\langle(d_{true} - A m)(d_{true}-Am)^T\rangle$. Knowing that $\langle (m-mt)(m-mt)^T\rangle$ is $(A^TN^{-1}A)^{-1}$, the error is $A(A^TN^{-1}A)^{-1}A^T$

```python
err = a@corr@a.transpose()
plt.imshow(err)
plt.colorbar()
```

**Assuming that the noise is uncorrelated in space (bold! I know!)**, we can histogram the estimated errors along the diagonal (variance, not covariance).

```python
var = np.diag(err)
plt.hist(np.sqrt(var), histtype = 'barstacked')
plt.title("Estimated $\sigma$ of data")
print("RMS error in data: ".ljust(10), f"{np.sqrt(np.sum(var)/len(var)):.3f}")
print("The relative error is small, max relative error = ", np.abs(np.sqrt(var)/z).max())
```

# Focal point
For a parabola (2D), the focal point formula is $y= \frac{x^2}{4f}$. 

For a paraboloid (3D), a similar equation can be found by setting $x$ or $y$ to zero (or taking any vertical plane across the origin). For example, $z-z_0 = a((x-x_0)^2)$.

Looking at the forms for parabola and paraboloid, we can tell that $f = \frac{1}{4a}$. 
## Error propagation:
$$\varepsilon(f) = |\frac{\partial f}{\partial a}\sigma_a| = |\frac{1}{4a^2}\sigma_a|$$

```python
f = 1/(4*aa)
ef = np.abs(1/(4*aa**2)*np.sqrt(np.diag(corr))[0])
print('Focal length is:'.ljust(10), f"{f:.1f}","+/-" ,f"{ef:.1f}")
```

**Our target focal length is 1.5 m. Assuming we are working with mm, then our target focal length falls within $2\sigma$ of our measurement.**

# Strategy (two parts):
* Introduce ellipticality into our formula. $$z-z0 = a(x-x0)^2 + b(y-y0)^2$$

* function with ability to rotate (x,y) by an angle $\theta$ from 0 to Pi/2. ($90^\circ$ is sufficient to align the two principle axis at least once.)

* Use $\chi^2$ as a measure, minimize $\chi^2$ with respect to $\theta$ which would give us the most reasonable fit to an **ellipsoid**.

With one extra degree of freedom, expansion leads to 
$$z=a x^2+b y^2-2 a x x_0-2 b y y_0+(b y_0^2+a x_0^2+z_0)\\
z=A x^2+B y^2 + Cx + Dy+E
$$

# Rotation
positive $\theta$ rotates (x,y) counter-clockwise on x-y plane (opposite to Problem Set instruction).

```python
def rot(x,y,theta):
    xp = np.cos(theta)*x - np.sin(theta)*y
    yp = np.sin(theta)*x + np.cos(theta)*y

    return xp,yp

def chisqr(theta):
    ## rotate original x,y data
    xx, yy = rot(x,y, theta)
    a = np.vstack([xx**2, yy**2, xx, yy, np.ones(len(xx))]).transpose()
    corr = np.linalg.inv(a.transpose()@a)
    m = corr@a.transpose()@z
    ## assuming our measurement uncertainty is uniform across all (x,y)
    chisqr = (a@m-z).transpose()@(a@m-z)
    
    return chisqr

chisqr = np.vectorize(chisqr)
```

```python
t = np.linspace(-np.pi, np.pi, 300)
plt.plot(t,chisqr(t)) 
plt.xlabel("Rotation Deg")
plt.ylabel("$\chi^2$ value")
plt.title('$\chi^2 vs \theta$ ')
```

**Indeed it does seem like there is two principle axis dividing the circular symmetry into four quadrants.**

```python
# Could use newton's method combined with our derivative function from Assignment 1 ... 
# But stealing runs in my blood...
from scipy.optimize import minimize 
```

```python
min_theta = minimize(chisqr,0, bounds = [(0,np.pi/2)]).x[0]
min_theta
```

**Now we can simply rotate, fit and extract the focal lengths (foci ;D) from fitted parameters**

```python
xx, yy = rot(x,y, min_theta)
a = np.vstack([xx**2, yy**2, xx, yy, np.ones(len(xx))]).transpose()
corr = np.linalg.inv(a.transpose()@a)
m = corr@a.transpose()@z
## assuming our measurement uncertainty is uniform across all (x,y)
chisqr = (a@m-z).transpose()@(a@m-z)
```

```python
print("chi squared value: ", chisqr)
```

```python
f1 = 1/(4*m[0])
f2 = 1/(4*m[1])
ef1 = np.abs(1/(4*m[0]**2)*np.sqrt(np.diag(corr))[0])
ef2 = np.abs(1/(4*m[1]**2)*np.sqrt(np.diag(corr))[1])
print("First focal length: ".rjust(30), f"{f1:.2f}", " +/- ", f"{ef1:.2f}")
print("Second focal length: ".rjust(30), f"{f2:.2f}", " +/- ", f"{ef2:.2f}")
print("Eccentricity: ".rjust(30), f"{np.sqrt(1- (f2/f1)**2):.2f}")
```
