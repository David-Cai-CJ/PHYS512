```python
import numpy as np
import matplotlib.pylab as plt
```

```python
def rk4_step(f,x,y,h):
    global count
    k1 = h*f(x,y)
    k2 = h*f(x+h/2,y+k1/2)
    k3 = h*f(x+h/2,y+k2/2)
    k4 = h*f(x+h,  y+k3)
    count +=4
    return y + (k1+2*k2+2*k3+k4)/6
```

```python
f1 = lambda x, y: y/(1 + x**2)
```

We use `count` variable to record the total number of function evaluations.

```python
x = np.linspace(-20, 20, 201)
h = np.median(np.diff(x))
y0 = -1
count = 0

# ---------
yy = np.zeros(len(x)-1)

y = y0

for i in np.arange(len(yy)):
    y = rk4_step(f1,x[i], y,h)
    yy[i] = y
    
yy = np.hstack([y0,yy])
print("Total number of function calls", count)
```

```python
plt.plot(x, yy, 'k-', label = "RK4 solution")
plt.legend()
fig = plt.gcf()
ax = plt.gca()
```

# Fourth order accurate
$$\varepsilon\sim O(h)^4$$
Taking two steps of size $\frac{h}{2}$, our fourth order error has a factor $\frac{1}{16}$ in front.
In comparison, taking one step of size $h$, the factor for leading order error is $1$.

To cancel this error, we **weight** the smaller step by $16$,  subtract the results and divide to normalize.
$$ y_{d} = \frac{16 y_{h/2, \text{twice}} - y_{h}}{15}$$

```python
def rk4_stepd(fun,x,y,h):
    y0 = y
    ## Two small steps
    y1 = rk4_step(fun, x, y0, h/2)
    y2 = rk4_step(fun, x + h/2, y1, h/2)
    ## One big step
    yH = rk4_step(fun, x, y0, h)
    
    return (16*y2 - yH)/15
```

# 12 Times per Call. Three times the effort.
Because we called rk4_step three times in `rk4_stepd` and each time `rk4_step` evaluates the function four times, every time `rk4_stepd` is called, we evaluate `fun` $12$ times.

To conserve the number of function evaluations (roughly), we take `np.ceil(200/3)`steps

```python
xd = np.linspace(-20, 20, int(np.ceil(201/3)))
hd = np.median(np.diff(xd))
y0 = -1
count = 0

# -------
yyd = np.zeros(len(xd)-1)

y = y0

for i in np.arange(len(yyd)):
    y = rk4_stepd(f1,xd[i], y,hd)
    yyd[i] = y
    
yyd = np.hstack([y0,yyd])
print("Total number of function calls", count)
```

```python
ax.plot(xd, yyd, 'r--', label = 'RK4d solution')
fig.legend()
fig
```

# Plotting difference vs. TRUTH
The solution to ODE is $y = c_0 \text{Exp}(\text{ArcTan}(x))$. Plug in $y_0 =-1, x0 = -20$ yields $c_0=-\text{Exp}(\text{ArcTan}(20))$.

```python
f = lambda x: -np.exp(np.arctan(20))*np.exp(np.arctan(x))
# True values 
tyy = f(x)    # RK4
tyyd = f(xd)  # RK4d


# Plotting
plt.clf()
plt.plot(x, np.abs(yy - tyy), 'r--', label = 'RK4 Difference')
plt.plot(xd, np.abs(yyd - tyyd), 'k--', label = 'RK4d Difference')
plt.ylabel('$\Delta y$ ')
plt.xlabel('$x$ ')
plt.yscale('log')
plt.legend()
```

```python
print('RMS error from RK4'.ljust(30),f"{np.sqrt(np.sum(np.abs(yy - tyy)**2)/len(x)):.2E}")
print('RMS error from RK4d'.ljust(30), f"{np.sqrt(np.sum(np.abs(yyd - tyyd)**2)/len(xd)):.2E}")
```

# Conclusion
Simply from the look of the plot above, I would say that the RK4d performs better than RK4 given the same number of function evaluations. The RMS error coming from `RK4d` is almost one order of magnitude lower.
